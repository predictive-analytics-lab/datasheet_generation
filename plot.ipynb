```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set_context("talk")
sns.set(rc={'figure.figsize':(14,12)}, font_scale=1.2, style="white")
sns.despine(left=True)
plt.rc('text', usetex=False)
colors = ["#E96832", '#b0d79b']
sns.set_palette(colors)
```

## Load Result CSV

```python
results = pd.read_csv("results.csv", index_col=0)
```

```python
fig, axs = plt.subplots(ncols=2, nrows=2, sharex=True)


p1 = sns.violinplot(x = "Algorithm", y = "Skeleton Precision", data=results, ax=axs[0, 0], scale='count', bw=.5)
p2 = sns.violinplot(x = "Algorithm", y = "Skeleton Recall", data=results, ax=axs[0, 1], scale='count', bw=.5)
p3 = sns.violinplot(x = "Algorithm", y = "V_Structure_Precision", data=results, ax=axs[1, 0], scale='count', bw=.5)
p4 = sns.violinplot(x = "Algorithm", y = "V_Structure_Recall", data=results, ax=axs[1, 1], scale='count', bw=.5)


p1.set_xlabel("Structure Learning Algorithm")
p2.set_xlabel("Structure Learning Algorithm")
p3.set_xlabel("Structure Learning Algorithm")
p4.set_xlabel("Structure Learning Algorithm")

p1.set_ylabel("Skeleton Precision")
p2.set_ylabel("Skeleton Recall")
p3.set_ylabel("V-Structure Precision")
p4.set_ylabel("V-Structure Recall")

p1.set_ylim([0, 1.1])
p2.set_ylim([0, 1.1])
p3.set_ylim([0, 1.1])
p4.set_ylim([0, 1.1])
```
